```python
import ast
df=pd.read_csv("averages_enrol.csv",converters={"ccode":ast.literal_eval})#,dtype={"ccode":list})
df[["term"]]=pd.Categorical(df.term)
df[["section"]]=pd.Categorical(df.section)
df
```

```python
### Predict course overall rating
from sklearn.linear_model import LinearRegression, LassoCV, Ridge, RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

# feats=["help","interesting","visual","oral","q_treatment","expl_lvl","assign_helpful"]
feats=list(df1.select_dtypes('number').columns)
[feats.remove(xx) for xx in ["salary","enrolled","overall"]]
X=df1[feats]
x1,x2,y1,y2=train_test_split(X,df1.overall,random_state=42)
ss=StandardScaler()
# x1=pd.DataFrame(ss.fit_transform(x1),columns=x1.columns)
# x2=ss.transform(x2)


# x1 = np.array(x1)
x1 = sm.add_constant(x1, prepend=False)
# lr=LinearRegression()
# lr=LassoCV(cv=10)
# lr=RidgeCV(alphas=10.**np.arange(-3,3))
lr=sm.OLS(y1,x1)
lr.fit().summary()
# lr.fit(x1,y1)
# print(f"train error: {np.mean(np.square(lr.predict(x1)-y1))}")
# print(f"test error:  {np.mean(np.square(lr.predict(x2)-y2))}")
```

```python
from sklearn.metrics import r2_score
r2_score(lr.predict(x2),y2)
```

```python
for feat,coef in sorted(zip(feats,lr.coef_),key=lambda x:abs(x[1]),reverse=True):
    print(f"{feat:<16} {coef:7.4f}") # lasso
```

# Interpretation of Linear Regression coefficients

Overall effectiveness rating correlates positively with:
* Be organized!
* Have good oral presentations
* Answer students' questions well

and to a lesser extent:
* Have good visual presentations
* Teach interesting courses
* Be available for help

Interesting notes:
* Results are similar regardless of regularization method (or even no regularization). Thus we can be confident about robustness.
* Organization is most important
* Oral far more important than visual
* Everything after first 6 features is pretty much unimportant (in the linear model)

```python
# TENURE PREDICTOR!!!
dfp=df1
# dfp=df1[~df1.title.isna()]
dfp=dfp.groupby("instructor").mean()
dfp["num_taught"]=df1.groupby("instructor").size()
dfp["title"]=df1.groupby("instructor").first()[["title"]]
# dfp=dfp.reset_index()
feats=list(df1.select_dtypes('number').columns)
[feats.remove(xx) for xx in ["salary","num_responses","enrolled"]]
has_tenure=dfp.title.isin(["Professor","Associate Professor"])

xc1,xc2,yc1,yc2=train_test_split(dfp[feats],has_tenure,test_size=.2,random_state=42)
ss=StandardScaler()
xc1=ss.fit_transform(xc1)
xc2=ss.transform(xc2)

from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
# lr2=LinearRegression()
# lr2=LassoCV(cv=5)
lr2=LogisticRegressionCV(penalty="l1",cv=5,solver='liblinear',max_iter=999,random_state=42)#penalty="l2",C=100,solver='lbfgs',max_iter=999)
lr2.fit(xc1,yc1)
print(f"train error: {np.mean(lr2.predict(xc1)!=yc1)}")
print(f"test error:  {np.mean(lr2.predict(xc2)!=yc2)}")
print(f"baseline:    {np.mean(has_tenure)}")
```
